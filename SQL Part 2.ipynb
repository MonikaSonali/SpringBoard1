# QUESTION 10
/* Q10: Produce a list of facilities with a total revenue less than 1000.
The output of facility name and total revenue, sorted by revenue. Remember
that there's a different cost for guests and members! */


```python
df_Q10=pd.read_sql_query("""with cte_mem (facid ,tot_mem_slots) as (
select facid , sum(slots) as tot_mem_slots
from bookings
where memid <> 0
group by facid 
) ,

 cte_guest (facid,tot_guest_slots)as(
select facid , sum(slots) as tot_guest_slots
from bookings 
where memid = 0 
group by facid 
)
select b.facid , f.name as facility ,tot_mem_slots ,tot_guest_slots,f.membercost as mem_cost , f.guestcost ,
 CASE WHEN b.memid= 0 THEN tot_mem_slots*f.membercost 
      ELSE tot_guest_slots*f.guestcost END as revenue 
from bookings as b
left join facilities as f on f.facid=b.facid
join cte_mem on b.facid = cte_mem.facid
join cte_guest on b.facid = cte_guest.facid
group by b.facid
order by revenue desc ;""", conn)
df_Q10
```

# QUESTION 12
/* Q12: Find the facilities with their usage by member, but not guests */

```python
df_Q12=pd.read_sql_query("""select m.firstname as FirstName, m.surname as Surname ,f.name as Facility , sum(b.slots) as total_slots from bookings as b
left join facilities as f on b.facid = f.facid
left join members as m on m.memid = b.memid
where m.memid <> 0 or m.firstname <> 'GUEST'
group by FirstName , Facility;
""",conn)
df_Q12['MemberName']=df_Q12['FirstName'] +' '+df_Q12['Surname']
# One slot is half an hour
df_Q12['UsageTime(hrs)']=df_Q12['total_slots']*0.5
df_Q12
```
