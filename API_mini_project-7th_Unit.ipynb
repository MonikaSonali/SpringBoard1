2. Convert the returned JSON object into a Python dictionary.

```python
url="https://www.quandl.com/api/v3/datasets/XFRA/AGB2/data.json?&start_date=2017-01-01&end_date=2017-12-31&api_key=sBazky2vUMD2RpLvdbgC"
r=requests.get(url)
json_data=r.json()
# print(json_data)
jsondict=dict(json_data)
print("\nDICT-----",jsondict['dataset_data']['data'],"-----",len(jsondict['dataset_data']['data']))

    
```

```python
 highNlow={'Highs':[],'Lows':[] ,'Openings':[] ,'Closings':[],'Volumes':[]}
for i in range(0,len(jsondict['dataset_data']['data'])):
    highNlow['Highs'].append(jsondict['dataset_data']['data'][i][2])
    highNlow['Lows'].append(jsondict['dataset_data']['data'][i][3])
    highNlow['Openings'].append(jsondict['dataset_data']['data'][i][1])
    highNlow['Closings'].append(jsondict['dataset_data']['data'][i][4])
    highNlow['Volumes'].append(jsondict['dataset_data']['data'][i][5])
```

```python
# print("highNlow  highNlow['Highs'] ------",highNlow['Highs']," \nhighNlow['Lows'] ------\n ",highNlow['Lows'],"\n---Openings",
#       highNlow['Openings'])
print("LENGTH OF HIGHS --- ",len(highNlow['Highs']),"\n LENGTH OF LOWS --- ",len(highNlow['Lows']))
```

```python
HighestPrice=max(highNlow['Highs'])
LowestPrice=min(highNlow['Lows'])
HighestOpeningPrice = max(highNlow['Openings'])
print("The Highest Opening Price is " , HighestOpeningPrice)
print("The Highest Price is " ,HighestPrice)
print("The Lowest  Price is " , LowestPrice)
```

```python
difference=[]
for high ,low in zip(highNlow['Highs'],highNlow['Lows']):
    difference.append(high-low)
# print("Changes per Day --- ",difference)
print("Largest change in a day is ", max(difference))
```

```python
closingDiff=[]
for i in range(1,len(highNlow['Closings'])):
    if i == 1:
         closingDiff.append(highNlow['Closings'][i] -highNlow['Closings'][0])
    else :
         closingDiff.append(highNlow['Closings'][i] -highNlow['Closings'][i-1])
print("LENGTH OF CLOSING DIFF" ,len(closingDiff) )                  
# print("Closing Differces --- ",closingDiff,"LENGTH OF CLOSING DIFF" ,len(closingDiff) )     
print("Largest change between any two days = ",max(closingDiff))
```
